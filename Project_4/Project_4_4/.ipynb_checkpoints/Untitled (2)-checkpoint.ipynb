## 3.

### Read data file and store object columns (Categorical Variables)

```python
car_data_path = '/home/raycatcher/Desktop/Data_Science_Project/car.data'
car_data = pd.read_csv(car_data_path)
car_data.replace('?', np.nan)
s = (car_data.dtypes == 'object')
# Make a list of them
object_cols = list(s[s].index)
# Make copy to avoid changing original data 
car_data
```

### Checking missing values

```python
for column in car_data.columns:
    missing = car_data[column].isna().sum()
    print(column+': '+str(missing))
```

### Splitting data into train and test sets

```python
from sklearn.model_selection import train_test_split
x = car_data.columns[:-1].tolist()
y = car_data.columns[-1:].tolist()
train_x, val_x, train_y, val_y = train_test_split(car_data[x], car_data[y], test_size=0.3, random_state=69)
```

### Label encoding using LabelEncoder class

```python
from sklearn.preprocessing import LabelEncoder
# Finding categorical variables
label_train_x = train_x.copy()
label_train_y = train_y.copy()
label_val_x = val_x.copy()
label_val_y = val_y.copy()

# Apply label encoder to each column with categorical data
label_encoder = LabelEncoder()
for col in object_cols[:-1]:
    label_train_x[col] = label_encoder.fit_transform(train_x[col])
    label_val_x[col] = label_encoder.transform(val_x[col])
label_train_y['class'] = label_encoder.fit_transform(train_y['class'])
label_val_y['class'] = label_encoder.transform(val_y['class'])
label_val_y.shape
```

### Train and evaluate Neural Network model, using MLPClassifier class

```python
from sklearn.neural_network import MLPClassifier
mlp = MLPClassifier(hidden_layer_sizes=(150,100,50), max_iter=300,activation = 'relu',solver='adam',random_state=1)
mlp.fit(label_train_x, label_train_y.values.ravel())
predict_train = mlp.predict(label_val_x)
#predict_test = mlp.predict(label_val_y)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(confusion_matrix(label_val_y,predict_train), end = '\n\n')
print(classification_report(label_val_y,predict_train))
```

### Train and evaluate Decision Tree model, using DecisionTreeRegressor class

```python
from sklearn.tree import DecisionTreeRegressor
dtr = DecisionTreeRegressor(random_state=1)
dtr.fit(label_train_x, label_train_y)
predict_train = dtr.predict(label_val_x)
print(confusion_matrix(label_val_y,predict_train), end = '\n\n')
print(classification_report(label_val_y,predict_train))
```

### Train and evaluate Naïve Bayes model, using DecisionTreeRegressor class

```python
from sklearn.naive_bayes import MultinomialNB
mnb = MultinomialNB()
mnb.fit(label_train_x, label_train_y.values.ravel())
predict_train = mnb.predict(label_val_x)
print(confusion_matrix(label_val_y,predict_train), end = '\n\n')
print(classification_report(label_val_y,predict_train))
```

# 4. Clustering

## Part 3

### Reading data file

```python
iris_path = '/home/raycatcher/Desktop/Data_Science_Project/iris.data'
iris_data = pd.read_csv(iris_path)
x = iris_data[iris_data.columns.tolist()[:-1]]
y_not_labeled = iris_data[iris_data.columns.tolist()[-1:]]
#y.reset_index(drop=True, inplace=True)
label_encoder = LabelEncoder()
y = y_not_labeled.copy()
y['class'] = label_encoder.fit_transform(y_not_labeled['class'])
y = y['class']
```

### Fit k-means model, using KMeans class

```python
from sklearn.cluster import KMeans
km_model = KMeans(n_clusters=5)
km_model.fit(x)
```

### Rand-score and nmi metrics for our k-means model

```python
from sklearn.metrics.cluster import adjusted_rand_score
from sklearn.metrics.cluster import normalized_mutual_info_score
rand_score = adjusted_rand_score(km_model.labels_, y)
nmi = normalized_mutual_info_score(km_model.labels_, y)
print(f'rand-score: {rand_score}')
print(f'nmi: {nmi}')
```

### Rand-score and nmi metrics for our fcm model

```python
from fcmeans import FCM
fcm_model = FCM(n_clusters=5)
fcm_model.fit(x)
fcm_labels = fcm_model.u.argmax(axis=1)
rand_score = adjusted_rand_score(fcm_labels, y)
nmi = normalized_mutual_info_score(fcm_labels, y)
print(f'rand-score: {rand_score}')
print(f'nmi: {nmi}')
```

## Part 4

### Reading "Spectf Heart" data file

```python
from sklearn import datasets
spectf_path = '/home/raycatcher/Desktop/Data_Science_Project/SPECT.train'
spectf_data = pd.read_csv(spectf_path)
x = spectf_data[spectf_data.columns.tolist()[:-1]]
y_not_labeled = spectf_data[spectf_data.columns.tolist()[-1:]]
y = y['class']
```

### Fit k-means model, using KMeans class

```python
from sklearn.cluster import KMeans
km_model = KMeans(n_clusters=2)
km_model.fit(x)
```

### rand-score, nmi, purity and ari metrics for our k-means model

```python
from sklearn.metrics.cluster import adjusted_rand_score
from sklearn.metrics.cluster import normalized_mutual_info_score
from sklearn import metrics
from sklearn.metrics.cluster import adjusted_rand_score
# Rand-score metrics
rand_score = adjusted_rand_score(km_model.labels_, y)
# NMI metrics
nmi = normalized_mutual_info_score(km_model.labels_, y)
# Purity metrics
confusion_matrix = metrics.cluster.contingency_matrix(km_model.labels_, y)
purity = np.sum(np.amax(confusion_matrix, axis=0)) / np.sum(confusion_matrix) 
# AIR metrics
ari = adjusted_rand_score(km_model.labels_, y)
# Printing results
print(f'Rand-score: {rand_score}')
print(f'NMI: {nmi}')
print(f'Purity: {purity}')
print(f'AIR: {ari}')
```

```python
from fcmeans import FCM
fcm_model = FCM(n_clusters=5)
fcm_model.fit(x)
fcm_labels = fcm_model.u.argmax(axis=1)
# Rand_score
rand_score = adjusted_rand_score(fcm_labels, y)
# NMI
nmi = normalized_mutual_info_score(fcm_labels, y)
# Purity
confusion_matrix = metrics.cluster.contingency_matrix(fcm_labels, y)
purity = np.sum(np.amax(confusion_matrix, axis=0)) / np.sum(confusion_matrix) 
# ARI
ari = adjusted_rand_score(fcm_labels, y)
# Printing results
print(f'rand-score: {rand_score}')
print(f'nmi: {nmi}')
print(f'Purity: {purity}')
print(f'AIR: {ari}')
```
