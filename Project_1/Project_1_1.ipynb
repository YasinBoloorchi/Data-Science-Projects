# Project 1 Exercise 2. UOK website scraping

```python
from bs4 import BeautifulSoup
import requests
```

## Get website source

```python
# get home page of University of Kurdistan website
home_page = requests.get('https://uok.ac.ir/fa.aspx')
home_page_souped = BeautifulSoup(home_page.text, 'html.parser')
```

## Get all news and set it's kinds

```python
# save news kind in a list (we can save it automaticaly but its' not necessary)
main_titles = ['اخبار ویژه', 'اخبار آموزشی', 'اخبار دانشجویی', 'اخبار پژوهشی', 'اخبار انتصابات',
'اخبار به زبان کردی', 'سایر اخبار']

# gather all news from website
all_news_stories = home_page_souped.findAll('div', attrs={'class':'NewsStories-Inner'})
print('news count: ',len(all_news_stories))
```

## Show all news with their requested informations

```python
# for each kind of news in the website
for main_count in range(7):
    
    # get the news kind name from main_title list
    news_kind = main_titles[main_count]
    print('-'*50)

    # all news in that kind
    news_stories = all_news_stories[main_count].findAll('div', attrs={'class':'NewsStory'})

    # for each news in this kind 
    for news_number in range(len(news_stories)):

        this_news = news_stories[news_number]

        # get the needed information
        news_title = this_news.findAll('h2')[0].text
        news_pub_date = this_news.findAll('div', attrs={'class':'PubDate'})[0].text
        news_text = this_news.findAll('p')[0].text
        news_link = 'http://uok.ac.ir' + this_news.findAll('a', attrs={'class':'More'})[0]['href']

        # print all gathered inforamtion
        print('\n* Title:', news_title)
        print('\n* Kind:', news_kind)
        print('\n* Pub date:', news_pub_date)
        print('\n* News: ', news_text)
        print('\n* Link:', news_link)
    
        print('-'*50)    
```
